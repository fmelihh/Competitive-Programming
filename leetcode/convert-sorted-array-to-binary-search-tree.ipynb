```python
# Definition for a binary tree node.

from typing import List, Optional

class TreeNode:
    def __init__(self, val=0, left=None, right=None, height: int = 1):
        self.val = val
        self.left = left
        self.right = right
        self.height = height
    
class Solution:
    def get_height(self, root: TreeNode) -> int:
        if not root:
            return 0

        return root.height

    def get_balance(self, root: TreeNode) -> int:
        if not root:
            return 0
        return self.get_height(root.left) - self.get_height(root.right)

    def right_rotate(self, z: TreeNode) -> TreeNode:
        y = z.left
        T3 = y.right

        y.right = z
        z.left = T3

        z.height = 1 + max(self.get_height(z.left), self.get_height(z.right))
        y.height = 1 + max(self.get_height(y.left), self.get_height(y.right))

        return y

    def left_rotate(self, z: TreeNode) -> TreeNode:
        y = z.right
        T2 = y.left

        y.left = z
        z.right = T2

        z.height = 1 + max(self.get_height(z.left), self.get_height(z.right))
        y.height = 1 + max(self.get_height(z.left), self.get_height(z.right))
        return y
    
    def insert(self, root: TreeNode, key: int) -> TreeNode:
        if not root:
            return TreeNode(val=key)
        elif key < root.val:
            root.left = self.insert(root.left, key)
        else:
            root.right = self.insert(root.right, key)

        root.height = 1 + max(self.get_height(root.left), self.get_height(root.right))
        balance = self.get_balance(root)

        if balance > 1 and key < root.left.val:
            return self.right_rotate(root)
            
        if balance < -1 and key > root.right.val:
            return self.left_rotate(root)
            
        if balance > 1 and key > root.left.val:
            root.left = self.left_rotate(root.left)
            return self.right_rotate(root)

        if balance < -1 and key < root.right.val:
            root.right = self.right_rotate(root.right)
            return self.left_rotate(root)
            
        return root
        
    def sortedArrayToBST(self, nums: List[int]) -> Optional[TreeNode]:
        node = None
        for num in nums:
            node = self.insert(node, num)
        return node
```
